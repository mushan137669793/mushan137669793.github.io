```python
from glob import glob
import torchaudio
from IPython.display import Audio, display
import os

def from_text(filename):
    with open(filename, "r") as f:
        text = f.readlines()
    return text

def to_text(filename, data):
    with open(filename, "w") as f:
        for d in data:
            f.write(d)
            
def from_pickle(filename):
    with open(filename, "rb") as f:
        data = pickle.load(f)
    return data
```

```python
hts = from_text("./ita_dan.html")
```

```python
nht = []

for h in hts:
    if 'preload="none"' in h:
        s = h.split('src="')[-1].split('"></audio>')[0]
        tar = f".{s.split('.')[1]}.{s.split('.')[-2]}.flac"
        nht.append(h.replace(s, tar))
        os.rename(s, tar)
    else:
        nht.append(h)
```
